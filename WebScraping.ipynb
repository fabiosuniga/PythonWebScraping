```python
import requests
from bs4 import BeautifulSoup

def buscar_pagina(url):
    # a aplicação finge ter um navegador proprio para acessar o link url
    resposta = requests.get(url)

    # verifica se conseguiu entrar no site
    if resposta.status_code == 200:
        print("Sucesso! Conseguimos entrar no site.\n")
        # retorna todo texto em HTML do site
        return resposta.text
    else:
        print(f"Deu ruim. Erro: {resposta.status_code}")
        return None

# testando a aplicação
url_alvo = "http://quotes.toscrape.com/"
html_da_pagina = buscar_pagina(url_alvo)


# habilita o código a entender o texto em HTML
sopa = BeautifulSoup(html_da_pagina, 'html.parser')

# todo texto que queremos está dentro de um 'div'
caixas_de_citacao = sopa.find_all('div', class_='quote')

# todo autor está inserido em um texto 'small', e toda citação em um texto 'span'
# criamos um dicionario e salvamos o autor e a frase dele
extrair_dados = lambda caixa: {
    "autor": caixa.find('small', class_='author').text,
    "frase": caixa.find('span', class_='text').text
}


# aplicamos um map convertido para lista, em todo texto 'div' do código, utilizando a regra onde usamos dicionário
lista_de_citacoes = list(map(extrair_dados, caixas_de_citacao))

# mostrando todas citações da página
for citacao in lista_de_citacoes:
    print(f"Autor: {citacao['autor']}\nFrase: {citacao['frase']}\n---")

# guardando somente citações do Albert Einstein
citacoes_einstein = list(filter(lambda citacao: citacao['autor'] == 'Albert Einstein', lista_de_citacoes))

# mostra quantas citações foram encontradas
print(f"Encontramos {len(citacoes_einstein)} frases do Albert Einstein na página!\n")

for item in citacoes_einstein:
    print(f"Frase: {item['frase']}\n---")

# definindo o nome do arquivo que vamos extrair da memória
nome_do_arquivo = "frases_einstein.txt"

# abre ou cria o arquivo, utilizando 'w' para sobrescrever e o enconding para evitar erros de acentuação
with open(nome_do_arquivo, 'w', encoding='utf-8') as arquivo:

    # cabeçalho do nosso arquivo
    arquivo.write("--- Citações de Albert Einstein ---\n\n")

    # inserindo cada item do nosso filtro
    for item in citacoes_einstein:
        arquivo.write(f"Frase: {item['frase']}\n")
        arquivo.write("---\n")

print(f"Sucesso! O arquivo '{nome_do_arquivo}' foi criado com as frases.")
```
